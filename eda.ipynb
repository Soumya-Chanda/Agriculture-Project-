```python
plt.style.use('seaborn-v0_8-darkgrid')  # Updated style name for compatibility

# Create line plots for temperature, rainfall, and crop yield over time
plt.figure(figsize=(14, 7))

# Average Temperature
plt.plot(data['Year'], data['Average_Temperature_C'], label='Average Temperature (°C)', marker='o')

# Add labels and title
plt.title('Trends Over Time: Temperature')
plt.xlabel('Year')
plt.ylabel('Values')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```
